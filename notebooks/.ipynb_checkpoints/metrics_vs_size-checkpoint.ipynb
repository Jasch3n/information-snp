```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.colors import Normalize
import matplotlib.cm as cm
import xarray as xr
import pandas as pd
from tqdm import tqdm
import warnings
import cartopy.crs as ccrs

import sys
sys.path.append("../scripts")
import VarMetrics, InfoTheoryMetrics
from InfoTheoryMetrics import I, optbins, I_jugaad
```

```python
from scipy.stats import iqr

def freedman_diaconis_bins(X):
    return int(np.ptp(X) / (2 * iqr(X) / np.power(len(X), 1/3)))
```

```python
from sklearn.feature_selection import mutual_info_regression

def calc_info(ensemble, signal, observation):
    N, t_len, lat_len, lon_len = ensemble.psl.shape
    
    # First dimension -- 0:Histogram, 1:KSG
    I_gf = np.zeros((2, lat_len, lon_len))
    I_sf = np.zeros((2, lat_len, lon_len))
    H_f = np.zeros((2, lat_len, lon_len))
    H_g = np.zeros((2, lat_len, lon_len))
    
    # Leave-One-Out array: 
    # First dimension -- as above
    # Second dimension: 0:I_sf, 1: H_f
    info_sf_LOO = np.zeros((2, 2, lat_len, lon_len))
    for lat in range(lat_len):
        for lon in range(lon_len):
            f_nt = ensemble.isel(lat=lat, lon=lon).psl.to_numpy()
            f_nt[np.isnan(f_nt)] = 0.0
            g_t = observation.isel(lat=lat, lon=lon).psl.to_numpy()
            s_t = signal.isel(lat=lat, lon=lon).psl.to_numpy()
            s_t[np.isnan(s_t)] = 0.0
            
#             f_bins = optbins(f_nt.flatten(), 100)
#             g_bins = optbins(g_t, 80)
#             s_bins = optbins(s_t, 80)
            f_bins = freedman_diaconis_bins(f_nt.flatten())
            g_bins = freedman_diaconis_bins(np.tile(g_t, N))
            s_bins = freedman_diaconis_bins(np.tile(s_t, N))
            
            # Histogram Calculations
            I_gf_hist, H_g_hist, H_f_hist = I_jugaad(g_t, f_nt, nbins_alt=(g_bins, f_bins))
            I_sf_hist, _, _ = I_jugaad(s_t, f_nt, nbins_alt=(s_bins, f_bins))
            I_gf[0,lat,lon] = I_gf_hist
            I_sf[0,lat,lon] = I_sf_hist
            H_g[0,lat,lon] = H_g_hist
            H_f[0,lat,lon] = H_f_hist
            
            # KSG Calculations
#             I_gf_ksg = mutual_info_regression(f_nt.flatten().T.reshape(-1, 1),
#                                               np.tile(g_t, N), 
#                                               n_neighbors=int(np.log(t_len)))[0]
#             I_sf_ksg = mutual_info_regression(f_nt.flatten().T.reshape(-1, 1),
#                                               np.tile(s_t, N), 
#                                               n_neighbors=int(np.log(t_len)))[0]
#             H_g_ksg = mutual_info_regression(g_t.T.reshape(-1, 1),
#                                              g_t, 
#                                              n_neighbors=int(np.log(t_len)))[0]
#             H_f_ksg = mutual_info_regression(f_nt.flatten().T.reshape(-1, 1),
#                                              f_nt.flatten(), 
#                                              n_neighbors=int(np.log(t_len)))[0]
#             I_gf[1,lat,lon] = I_gf_ksg
#             I_sf[1,lat,lon] = I_sf_ksg
#             H_g[1,lat,lon]= H_g_ksg
#             H_f[1,lat,lon] = H_f_ksg
            
            # ================ AVERAGE LEAVE-ONE-OUT CALCULATION ================
            I_sf_LOO_hist = 0.0
            I_sf_LOO_kraskov = 0.0
            for k in range(len(ensemble.n)):
                LOO_f = ensemble.where(ensemble["n"]!=k, drop=True).isel(lat=lat, lon=lon).psl.to_numpy()
                member = ensemble.isel(n=k, lat=lat, lon=lon).psl.to_numpy()
                LOO_f[np.isnan(LOO_f)] = 0.0
                member[np.isnan(member)] = 0.0
                
                # Histogram LOO calculation
                LOO_member_bins = freedman_diaconis_bins(np.tile(member, N-1))
                LOO_f_bins = freedman_diaconis_bins(LOO_f.flatten())
                I_sf_LOO_hist, _, H_LOO_f_hist = I_jugaad(member, LOO_f, nbins_alt=(LOO_member_bins, LOO_f_bins))
                info_sf_LOO[0,0,lat,lon] = I_sf_LOO_hist
                info_sf_LOO[0,1,lat,lon] = H_LOO_f_hist
                
                # KSG LOO calculation
#                 I_sf_LOO_ksg = mutual_info_regression(LOO_f.flatten().T.reshape(-1,1), 
#                                                       np.tile(member, LOO_f.shape[0]),
#                                                       n_neighbors=int(np.log(t_len)))[0]
#                 H_LOO_f_ksg = mutual_info_regression(LOO_f.flatten().T.reshape(-1,1), 
#                                                      LOO_f.flatten(),
#                                                      n_neighbors=int(np.log(t_len)))[0]
#                 info_sf_LOO[1,0,lat,lon] = I_sf_LOO_ksg
#                 info_sf_LOO[1,1,lat,lon] = H_LOO_f_ksg
            
    return I_gf, I_sf, H_g, H_f, info_sf_LOO
```

```python
from sklearn.feature_selection import mutual_info_regression

def calc_info_2(ensemble, signal, observation):
    """ Focus on one grid point only """
    N, t_len = ensemble.psl.shape
    
    # First dimension -- 0:Histogram, 1:KSG
    I_gf = np.zeros((2, lat_len, lon_len))
    I_sf = np.zeros((2, lat_len, lon_len))
    H_f = np.zeros((2, lat_len, lon_len))
    H_g = np.zeros((2, lat_len, lon_len))
    
    # Leave-One-Out array: 
    # First dimension -- as above
    # Second dimension: 0:I_sf, 1: H_f
    info_sf_LOO = np.zeros((2, 2, lat_len, lon_len))
    for lat in range(lat_len):
        for lon in range(lon_len):
            f_nt = ensemble.isel(lat=lat, lon=lon).psl.to_numpy()
            f_nt[np.isnan(f_nt)] = 0.0
            g_t = observation.isel(lat=lat, lon=lon).psl.to_numpy()
            s_t = signal.isel(lat=lat, lon=lon).psl.to_numpy()
            s_t[np.isnan(s_t)] = 0.0
            
            f_bins = freedman_diaconis_bins(f_nt.flatten())
            g_bins = freedman_diaconis_bins(np.tile(g_t, N))
            s_bins = freedman_diaconis_bins(np.tile(s_t, N))
            
            # Histogram Calculations
            I_gf_hist, H_g_hist, H_f_hist = I_jugaad(g_t, f_nt, nbins_alt=(g_bins, f_bins))
            I_sf_hist, _, _ = I_jugaad(s_t, f_nt, nbins_alt=(s_bins, f_bins))
            I_gf[0,lat,lon] = I_gf_hist
            I_sf[0,lat,lon] = I_sf_hist
            H_g[0,lat,lon] = H_g_hist
            H_f[0,lat,lon] = H_f_hist
            
            # ================ AVERAGE LEAVE-ONE-OUT CALCULATION ================
            I_sf_LOO_hist = 0.0
            I_sf_LOO_kraskov = 0.0
            for k in range(len(ensemble.n)):
                LOO_f = ensemble.where(ensemble["n"]!=k, drop=True).isel(lat=lat, lon=lon).psl.to_numpy()
                member = ensemble.isel(n=k, lat=lat, lon=lon).psl.to_numpy()
                LOO_f[np.isnan(LOO_f)] = 0.0
                member[np.isnan(member)] = 0.0
                
                # Histogram LOO calculation
                LOO_member_bins = freedman_diaconis_bins(np.tile(member, N-1))
                LOO_f_bins = freedman_diaconis_bins(LOO_f.flatten())
                I_sf_LOO_hist, _, H_LOO_f_hist = I_jugaad(member, LOO_f, nbins_alt=(LOO_member_bins, LOO_f_bins))
                info_sf_LOO[0,0,lat,lon] = I_sf_LOO_hist
                info_sf_LOO[0,1,lat,lon] = H_LOO_f_hist
            
    return I_gf, I_sf, H_g, H_f, info_sf_LOO
```

# Calculate the RPC metrics vs. size of Ensemble

```python
# Load in the datasets
f = xr.open_dataset("../data/ensembles/large_psl_decadal_ensemble_no_extrapolate.nc").drop_dims(["bnds"])
obs = xr.open_dataset("../data/obs/hadslp2/hadslp2_monthly_1850_2004.nc")
obs = obs.where(obs["time.year"]>=1962, drop=True)
obs = obs.where(obs["time.year"]<1972, drop=True)
obs = obs.sortby("lon")

# Fine-grain the observations to match the model outputs
with warnings.catch_warnings(action="ignore"):
    obs["time"] = obs.indexes["time"].to_datetimeindex()
obs = obs.interp_like(f, kwargs={"fill_value":"extrapolate"})
# f = f.interp_like(obs, kwargs={"fill_value":"extrapolate"})

# Linearly detrend the data
obs_lin_trend = xr.polyval(obs["time"], obs.psl.polyfit(dim='time', deg=1).polyfit_coefficients)
obs["psl"] = obs["psl"] - obs_lin_trend

f_lin_trend = xr.polyval(f["time"], f.psl.polyfit(dim='time', skipna=True, deg=1).polyfit_coefficients)
f["psl"] = f["psl"] - f_lin_trend

s = f.mean("n")
```

```python
lon_l, lon_r = 50, 70
lat_l, lat_r = -10, 10
sub_lon = np.where(np.logical_and(s.lon>50, s.lon<70))[0]
sub_lat = np.where(np.logical_and(s.lat>-10, s.lat<10))[0]
np.random.seed(2024)
regional_ensemble = f.isel(lon=sub_lon, lat=sub_lat)
regional_obs = obs.isel(lon=sub_lon, lat=sub_lat)
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([-180, 180, -90, 90], crs=ccrs.PlateCarree())
ax.coastlines(resolution='110m', linewidth=.5, color='black')
gl = ax.gridlines(draw_labels=True, linewidth=0.5, alpha=0.4, color='k',
                  linestyle='--')
gl.top_labels=False
gl.right_labels=False
ax.add_patch(patches.Rectangle(xy=[lon_l, lat_l], width=lon_r-lon_l, height=lat_r-lat_l,
                                facecolor='none', edgecolor='r',
                                transform=ccrs.PlateCarree()))
plt.title("Region of Focus")
plt.show()
```

```python
SAMPLE_SIZE = 25
STEP_SIZE = 2
MAX_SUBENSEMBLE_SIZE = 50

ensemble_sizes = np.arange(2, MAX_SUBENSEMBLE_SIZE, STEP_SIZE)

var_metrics_vs_size = np.zeros((3, SAMPLE_SIZE, len(ensemble_sizes)))
info_metrics = []
for i in range(len(ensemble_sizes)):
    for sample in tqdm(range(SAMPLE_SIZE), desc=f'Drawing subensembles of size {ensemble_sizes[i]}'):
        subensemble = regional_ensemble.isel(n=np.random.randint(0, len(f.n), ensemble_sizes[i]))
        subsignal = subensemble.mean("n")
        rho_o = VarMetrics.obs_pc(subsignal, regional_obs)
        snr_rho_m = VarMetrics.model_pc(subensemble, subsignal, intrinsic_method='snr')
        LOO_rho_m = VarMetrics.model_pc(subensemble, subsignal, intrinsic_method='leave_one_out')
        var_metrics_vs_size[:, sample, i] = [rho_o.mean(), snr_rho_m.mean(), LOO_rho_m.mean()]
        
        # I_arr = I_gf, I_sf, H_g, H_f, info_sf_LOO
        I_arr = calc_info(subensemble, subsignal, regional_obs)
        info_metrics.append(I_arr)
```

```python
n_metric, lat_len, lon_len = info_metrics[0][0].shape
info_metrics_np = np.zeros((4, len(ensemble_sizes), SAMPLE_SIZE, n_metric, lat_len, lon_len))
LOO_info_metrics_np = np.zeros((2, len(ensemble_sizes), SAMPLE_SIZE, n_metric, lat_len, lon_len))
for i in range(len(info_metrics)):
    I_gf, I_sf, H_g, H_f, info_sf_LOO = info_metrics[i]
    I_sf_LOO = info_sf_LOO[:,0,:,:]
    H_f_LOO = info_sf_LOO[:,1,:,:]
    info_metrics_np[0,int(i/10),i%10,:,:,:] = I_gf
    info_metrics_np[1,int(i/10),i%10,:,:,:] = I_sf
    info_metrics_np[2,int(i/10),i%10,:,:,:] = H_g
    info_metrics_np[3,int(i/10),i%10,:,:,:] = H_f
    LOO_info_metrics_np[0,int(i/10),i%10,:,:,:] = I_sf_LOO
    LOO_info_metrics_np[1,int(i/10),i%10,:,:,:] = H_f_LOO    
```

```python
mean_var_metrics = np.mean(var_metrics_vs_size, axis=1)
std_var_metrics = np.std(var_metrics_vs_size, axis=1)

mean_info = np.mean(info_metrics_np, axis=(2,4,5))
std_info = np.std(np.mean(info_metrics_np, axis=(4,5)), axis=2)

mean_LOO_info = np.mean(LOO_info_metrics_np, axis=(2,4,5))
std_LOO_info = np.std(np.mean(LOO_info_metrics_np, axis=(4,5)), axis=2)
```

```python
# Unpacking the variance metrics
rho_o = mean_var_metrics[0,:]
rho_m = mean_var_metrics[1,:]
LOO_rho_m = mean_var_metrics[2,:]

# Unpacking the information metrics
I_gf = mean_info[0,:,:]
I_sf = mean_info[1,:,:]
H_g = mean_info[2,:,:]
H_f = mean_info[3,:,:]
I_sf_LOO = mean_LOO_info[0,:,:]
H_f_LOO = mean_LOO_info[1,:,:]

gam_o_hist = I_gf[:,0] / H_g[:,0]
gam_m_hist = I_sf[:,0] / H_f[:,0]
LOO_gam_m_hist = I_sf_LOO[:,0] / H_f_LOO[:,0]

gam_o_ksg = I_gf[:,1] / H_g[:,1]
gam_m_ksg = I_sf[:,1] / H_f[:,1]
LOO_gam_m_ksg = I_sf_LOO[:,1] / H_f_LOO[:,1]

lam_o_hist = np.sqrt(1 - np.exp(-2*I_gf[:,0]))
lam_m_hist = np.sqrt(1 - np.exp(-2*I_sf[:,0]))
lam_o_ksg = np.sqrt(1 - np.exp(-2*I_gf[:,1]))
lam_m_ksg = np.sqrt(1 - np.exp(-2*I_sf[:,1]))
LOO_lam_m_hist = np.sqrt(1 - np.exp(-2*I_sf_LOO[:,0]))
LOO_lam_m_ksg = np.sqrt(1 - np.exp(-2*I_sf_LOO[:,1]))
```
