### Create a large `xarray` dataset object from multiple `.nc` files to represent an ensemble of models

```python
import numpy as np 
import matplotlib.pyplot as plt
import xarray as xr
import nc_time_axis
import pandas as pd
import re
import cftime
from tqdm import tqdm
import os
```

```python
# Directory containing all datasets that go into one ensemble
DATA_DIR = "../data/models/psl"
```

```python
# Use one of the datasets to homogenize the rest of them
hadcm3_ds = xr.open_dataset("../data/hadcm3_decadal_1966_1/psl_Amon_HadCM3_decadal1966_r8i3p1_196611-197612.nc")
```

```python
# use these coordinates to reduce/enhance the models to the same spatial resolution
fine_lon = np.linspace(hadcm3_ds.lon[0], hadcm3_ds.lon[-1], hadcm3_ds.sizes["lon"])
fine_lat = np.linspace(hadcm3_ds.lat[0], hadcm3_ds.lat[-1], hadcm3_ds.sizes["lat"])
unify_time = pd.date_range(start="1967-01-16", end="1977-01-16", freq="ME")

interp_data = []

# clean the datasets and prepare them for concatenation
for datafile in os.listdir(DATA_DIR):
    print(datafile)
    ds_temp = xr.open_dataset(os.path.join(DATA_DIR, datafile))
    # make sure the datasets are the same length along the time axis
    ds_temp = ds_temp.where(ds_temp["time.year"] >= 1967, drop=True) 
    # ds_temp = ds_temp.assign_coords(time=unify_time)

    # homogenize time object data types
    # if(type(ds_temp.indexes['time']) == xr.coding.cftimeindex.CFTimeIndex): 
    #     datetimeindex = ds_temp.indexes['time'].to_datetimeindex(unsafe=True)
    #     ds_temp["time"] = datetimeindex
    # else:
    #     ds_temp["time"] = ds_temp.indexes["time"]

    # datasets.append(ds_temp) # keep a copy of the original for reference
    ds_temp = ds_temp.interp(lat=fine_lat, lon=fine_lon, kwargs={"fill_value":"extrapolate"})
    interp_data.append(ds_temp.psl.to_numpy())
    
```

```python
f_data = np.stack(interp_data, axis=0)

# ensemble
f = xr.Dataset(
    data_vars = dict(psl=(["n", "time", "lat", "lon"], f_data)),
    coords = dict(
        n = ("n", np.arange(len(interp_data))),
        time=("time", unify_time),
        lon=("lon",fine_lon),
        lat=("lat",fine_lat)
    )
)

# signal
s = f.mean("n")

# noise
eta = f - s
```

```python
# save the dataset to an nc file
f.to_netcdf("../data/ensembles/psl_decadal_ensemble.nc")
```

```python
import warnings

MODEL_DIR = "../data/models"
ds_paths = []
for model in os.listdir(MODEL_DIR):
    # if model=="GFDL-CM2p1":
    #     continue
    # print(model)
    for member in os.listdir(os.path.join(MODEL_DIR, model, "psl")):
        ds_paths.append(os.path.join(MODEL_DIR, model, "psl", member))

# Add all the datasets into a list for concatenation
ds_list = []
interp_ds_list = []
with warnings.catch_warnings(action="ignore"):
    # Use the example dataset to interpolate every other
    example_ds = xr.open_dataset(ds_paths[0])
    example_ds['time'] = example_ds.indexes['time'].to_datetimeindex()
    
    for i in tqdm(range(len(ds_paths)), desc=""):
        ds = xr.open_dataset(ds_paths[i])
        ds_list.append(ds)
        # For the GFDL-CM2.1 model, drop the variables that we don't need
#         ds = ds.drop_vars(["lon_bnds", "lat_bnds", "time_bnds"])
        
#         psl_da = ds.isel(bnds=0).psl.interp_like(example_ds.psl)
        if "average_DT" in ds.variables:
            ds = ds.drop_vars(["average_DT", "average_T1", "average_T2"])
        
        # Suppress the runtime warnings against calendar conversions
        if type(ds.indexes['time'])!= pd.core.indexes.datetimes.DatetimeIndex:
            ds['time'] = ds.indexes['time'].to_datetimeindex()
        else:
            ds['time'] = ds.indexes['time']
        ds_new = ds.interp_like(example_ds, kwargs={"fill_value":0.0})
        interp_ds_list.append(ds_new)
        
ensemble = xr.concat(interp_ds_list, dim="n")
```
