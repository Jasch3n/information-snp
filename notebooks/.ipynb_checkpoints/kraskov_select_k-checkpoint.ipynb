```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
import matplotlib.cm as cm
import xarray as xr
import pandas as pd
from tqdm import tqdm
import warnings
import cartopy.crs as ccrs
from sklearn.feature_selection import mutual_info_regression

from matplotlib import rc
# rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
# rc('text', usetex=True)
plt.rcdefaults()

import sys 
sys.path.append("../scripts")
from InfoTheoryMetrics import I, optbins
```

```python
# Load in the datasets
f = xr.open_dataset("../data/ensembles/large_psl_decadal_ensemble_no_extrapolate.nc").drop_dims(["bnds"])
obs = xr.open_dataset("../data/obs/hadslp2/hadslp2_monthly_1850_2004.nc")
obs = obs.where(obs["time.year"]>=1962, drop=True)
obs = obs.where(obs["time.year"]<1972, drop=True)
obs = obs.sortby("lon")

# Fine-grain the observations to match the model outputs
with warnings.catch_warnings(action="ignore"):
    obs["time"] = obs.indexes["time"].to_datetimeindex()
obs = obs.interp_like(f, kwargs={"fill_value":"extrapolate"})
# f = f.interp_like(obs, kwargs={"fill_value":"extrapolate"})

# Linearly detrend the data
obs_lin_trend = xr.polyval(obs["time"], obs.psl.polyfit(dim='time', deg=1).polyfit_coefficients)
obs["psl"] = obs["psl"] - obs_lin_trend

f_lin_trend = xr.polyval(f["time"], f.psl.polyfit(dim='time', skipna=True, deg=1).polyfit_coefficients)
f["psl"] = f["psl"] - f_lin_trend

s = f.mean("n")
```

```python
def extract_helper(signal, observation, lat, lon):
    s_xy = signal.isel(lat=lat, lon=lon).psl.to_numpy()
    g_xy = observation.isel(lat=lat, lon=lon).psl.to_numpy()
    return s_xy, g_xy
```

```python
# Inspect a particular location
lat, lon = 60,15
# lat, lon = 30, 21
# lat, lon = 14,0
t = f.time.to_numpy().astype("datetime64[m]")
f_xy = f.isel(lat=lat, lon=lon).psl.to_numpy()
model_spread = np.nanstd(f_xy, axis=0)
model_spread = model_spread[:T_STOP]
s_xy, g_xy = extract_helper(s,obs,lat,lon)

fig = plt.figure()
plt.plot(t[:T_STOP], s_xy[:T_STOP], linewidth=5, label='Ensemble Mean')
plt.fill_between(t[:T_STOP], s_xy[:T_STOP]-model_spread/2, s_xy[:T_STOP]+model_spread/2, interpolate=True, alpha=0.2)
plt.plot(t[:T_STOP], g_xy[:T_STOP], linewidth=5, label='HadSLP2')
plt.legend()
plt.ylabel("Sea-Level Pressure Anomaly")
fig.set_figwidth(8)

lat_num = np.round(f.lat[lat].to_numpy(), 2)
lon_num = np.round(f.lon[lon].to_numpy(), 2)
plt.title(f"Lon: ${lon_num}^\circ$E, Lat: ${lat_num}^\circ${'N' if lat_num>0 else 'S'}");
```

```python
ks = np.arange(1,50)
s_xy_1, g_xy_1 = extract_helper(s,obs,30,21)
# s_xy_2, g_xy_2 = extract_helper(s,obs,14,0)
s_xy_2, g_xy_2 = extract_helper(s,obs,3,15)
rho_sg_1 = np.corrcoef(s_xy_1, g_xy_1)[0,1]
rho_sg_2 = np.corrcoef(s_xy_2, g_xy_2)[0,1]
I_sg_hist_1 = I(s_xy_1, g_xy_1, nbins_alt=(optbins(s_xy_1, 10), optbins(g_xy_1, 10)))[0]
I_sg_hist_2 = I(s_xy_2, g_xy_2, nbins_alt=(optbins(s_xy_2, 20), optbins(g_xy_2, 20)))[0]
I_sg_k_1 = np.zeros(len(ks))
I_sg_k_2 = np.zeros(len(ks))
for i in range(len(ks)):
    k = ks[i]
    I_sg_1 = mutual_info_regression(s_xy_1.T.reshape(-1,1), g_xy_1, n_neighbors=k)[0]
    I_sg_2 = mutual_info_regression(s_xy_2.T.reshape(-1,1), g_xy_2, n_neighbors=k)[0]
    I_sg_k_1[i] = I_sg_1
    I_sg_k_2[i] = I_sg_2
print(I_sg_hist_1, np.mean(I_sg_k_1))
print(I_sg_hist_2, np.mean(I_sg_k_2))
```

```python
fig, axs = plt.subplots(2,2)
lam_1 = np.sqrt(1-np.exp(-2*I_sg_k_1))
diff_lam_1 = np.gradient(lam_1)
axs[0,0].plot(ks, lam_1, label='Positive Correlation')
axs[0,0].plot(ks, np.ones(len(ks))*np.sqrt(1-np.exp(-2*I_sg_hist_1)), '--', label=r'$\rho_1$')
axs[1,0].plot(ks, diff_lam_1, label='Diff')

lam_2 = np.sqrt(1-np.exp(-2*I_sg_k_2))
diff_lam_2 = np.gradient(lam_2)
axs[0,1].plot(ks, lam_2, label='Negative Correlation')
axs[0,1].plot(ks, np.ones(len(ks))*np.sqrt(1-np.exp(-2*I_sg_hist_2)), '--', label=r'$\rho_2$')
axs[1,1].plot(ks, diff_lam_2, label='Diff')

data = [
    [lam_1, lam_2],
    [diff_lam_1, diff_lam_2]
]
for i, row in enumerate(axs):
    for j, ax in enumerate(row):
        ax.plot(np.ones(10)*np.round(np.log(len(s_xy_1))), 
                np.linspace(np.min(data[i][j]),np.max(data[i][j]),10), 
                '--', label=r'$k=\lceil\log(N)\rceil$')
        ax.legend(fontsize=5)
        ax.set_xlabel("$k$")

plt.tight_layout()
```

A smaller value of $k$ leads to a less biased estimate while a larger value of $k$ leads to more variance in the estimates. There is no universal rule for choosing $k$, and here we use $k=\log(N)$.
