```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
import matplotlib.cm as cm
import xarray as xr
import pandas as pd
from tqdm import tqdm
import warnings
import cartopy.crs as ccrs
from sklearn.feature_selection import mutual_info_regression

from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
# plt.rcdefaults()

import sys 
sys.path.append("../scripts")
from InfoTheoryMetrics import I, optbins, H, I_jugaad
import VarMetrics
```

```python
# Load in the datasets
f = xr.open_dataset("../data/ensembles/large_psl_decadal_ensemble_no_extrapolate.nc").drop_dims(["bnds"])
obs = xr.open_dataset("../data/obs/hadslp2/hadslp2_monthly_1850_2004.nc")
obs = obs.where(obs["time.year"]>=1962, drop=True)
obs = obs.where(obs["time.year"]<1972, drop=True)
obs = obs.sortby("lon")

# Fine-grain the observations to match the model outputs
with warnings.catch_warnings(action="ignore"):
    obs["time"] = obs.indexes["time"].to_datetimeindex()
obs = obs.interp_like(f, kwargs={"fill_value":"extrapolate"})
# f = f.interp_like(obs, kwargs={"fill_value":"extrapolate"})

# Linearly detrend the data
obs_lin_trend = xr.polyval(obs["time"], obs.psl.polyfit(dim='time', deg=1).polyfit_coefficients)
obs["psl"] = obs["psl"] - obs_lin_trend

f_lin_trend = xr.polyval(f["time"], f.psl.polyfit(dim='time', skipna=True, deg=1).polyfit_coefficients)
f["psl"] = f["psl"] - f_lin_trend

s = f.mean("n")
```

```python
xx, yy = 21, 32
np.random.seed(300)
s_partial_t = f.isel(n=np.random.choice(f.n, 10), lon=xx, lat=yy).mean("n").psl.to_numpy()
s_partial_t_2 = f.isel(n=np.random.choice(f.n, 5), lon=xx, lat=yy).mean("n").psl.to_numpy()
s_t = s.isel(lon=xx, lat=yy).psl.to_numpy()
g_t = obs.isel(lon=xx, lat=yy).psl.to_numpy()
```

```python
plt.scatter(s_t, g_t, marker='o', alpha=0.3)
plt.xlabel("$s$")
plt.ylabel("$g$")

plt.scatter(s_partial_t, g_t, marker='x', alpha=0.3)
plt.xlabel("$s$")
plt.ylabel("$g$")

plt.scatter(s_partial_t_2, g_t, marker='x', alpha=0.3)
plt.xlabel("$s$")
plt.ylabel("$g$")

print(np.round(np.corrcoef(s_t,g_t)[0,1],3))
print(np.round(np.corrcoef(s_partial_t,g_t)[0,1],3))
print(np.round(np.corrcoef(s_partial_t_2,g_t)[0,1],3))
```

```python
NPTS = 6000
rho_m = np.linspace(0,5,NPTS)
expected_rho_o = rho_m/np.sqrt(1 + rho_m**2)
plt.plot(rho_m, expected_rho_o, color='black')
plt.plot(rho_m, np.ones(len(rho_m)), '--', color='black')
plt.fill_between(rho_m[:NPTS//5], np.zeros(NPTS//5), expected_rho_o[:NPTS//5], color='steelblue', alpha=0.5)
plt.fill_between(rho_m[NPTS//5:], np.ones(NPTS-NPTS//5), np.zeros(NPTS-NPTS//5), color='steelblue', alpha=0.5)
plt.fill_between(rho_m[:NPTS//5], np.ones(NPTS//5), expected_rho_o[:NPTS//5], color='salmon', alpha=0.5)
plt.xlabel(r"$\rho_m$")
plt.ylabel(r"Expected $\rho_o$")
plt.xlim([0,5])
plt.ylim([0,1.01])
```
